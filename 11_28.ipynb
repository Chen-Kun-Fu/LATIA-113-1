```python
import pandas as pd
import numpy as np


df = pd.read_csv('A_stat_data_112.csv')
Humanities_and_history_df = df[df['mission_field'] == '人文史地']

difficulty_mapping = {'易': 1, '中': 2, '難': 3}
Humanities_and_history_df['mission_difficulty'] = Humanities_and_history_df['mission_difficulty'].map(difficulty_mapping)

Humanities_and_history_df = Humanities_and_history_df.sort_values(by='mission_status', ascending=True).reset_index(drop=True)

hard_Humanities_and_history_df = Humanities_and_history_df[Humanities_and_history_df['mission_difficulty'] == 3]
mid_Humanities_and_history_df = Humanities_and_history_df[Humanities_and_history_df['mission_difficulty'] == 2]
easy_Humanities_and_history_df = Humanities_and_history_df[Humanities_and_history_df['mission_difficulty'] == 1]

hardA_status_Humanities_and_history_df = hard_Humanities_and_history_df[hard_Humanities_and_history_df['mission_status'] == '已派發任務']
hardZ_status_Humanities_and_history_df = hard_Humanities_and_history_df[hard_Humanities_and_history_df['mission_status'] == '完成任務']
hardQ_status_Humanities_and_history_df = hard_Humanities_and_history_df[hard_Humanities_and_history_df['mission_status'] == '取消派發']

midA_status_Humanities_and_history_df = mid_Humanities_and_history_df[mid_Humanities_and_history_df['mission_status'] == '已派發任務']
midZ_status_Humanities_and_history_df = mid_Humanities_and_history_df[mid_Humanities_and_history_df['mission_status'] == '完成任務']
midQ_status_Humanities_and_history_df = mid_Humanities_and_history_df[mid_Humanities_and_history_df['mission_status'] == '取消派發']

easyA_status_Humanities_and_history_df = easy_Humanities_and_history_df[easy_Humanities_and_history_df['mission_status'] == '已派發任務']
easyZ_status_Humanities_and_history_df = easy_Humanities_and_history_df[easy_Humanities_and_history_df['mission_status'] == '完成任務']
easyQ_status_Humanities_and_history_df = easy_Humanities_and_history_df[easy_Humanities_and_history_df['mission_status'] == '取消派發']


hardA_num_rows = hardA_status_Humanities_and_history_df.shape[0]
hardZ_num_rows = hardZ_status_Humanities_and_history_df.shape[0]
hardQ_num_rows = hardQ_status_Humanities_and_history_df.shape[0]

midA_num_rows = midA_status_Humanities_and_history_df.shape[0]
midZ_num_rows = midZ_status_Humanities_and_history_df.shape[0]
midQ_num_rows = midQ_status_Humanities_and_history_df.shape[0]

easyA_num_rows = easyA_status_Humanities_and_history_df.shape[0]
easyZ_num_rows = easyZ_status_Humanities_and_history_df.shape[0]
easyQ_num_rows = easyQ_status_Humanities_and_history_df.shape[0]


summary_df = pd.DataFrame({
    'Status': ['名稱','派發', '完成', '取消','難度'],
    '難': ['人文史地難',hardA_num_rows, hardZ_num_rows, hardQ_num_rows,3],
    '中': ['人文史地中',midA_num_rows, midZ_num_rows, midQ_num_rows,2],
    '易': ['人文史地易',easyA_num_rows, easyZ_num_rows, easyQ_num_rows,1]
 })

Humanities_and_history_df = summary_df.transpose()
Humanities_and_history_df

Humanities_and_history_df.to_excel('Humanities_and_history_df.xlsx', index=False)
```
